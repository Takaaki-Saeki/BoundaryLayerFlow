```python
# モジュールのインポート
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 速度データの読み込み
data_vel = pd.read_csv('output_vel.csv', header=None)
```

```python
# 速度データの可視化
x_vel = data_vel[0]
y_vel = data_vel[1]

plt.plot(x_vel, y_vel)
plt.xlabel('distance (m)')
plt.ylabel('velosity (m/s)')
plt.title('Visualization of velocity')
plt.savefig('velocity.jpg')
```

```python
# 境界層暑さデータの読み込み
data_blt = pd.read_csv('output_blt.csv', header=None)
```

```python
# 境界層厚さデータの可視化
x_blt = data_blt[0]
y1 = data_blt[1]
y2 = data_blt[2]
y3 = data_blt[3]

plt.plot(x_blt, y1, 'g-')
plt.plot(x_blt, y2, 'r-')
plt.plot(x_blt, y3, 'b-')
plt.xlabel('distance (m)')
plt.ylabel('boundary layer thickness (m)')
plt.title('Visualization of boundary layer thickness')
plt.savefig('boundary_layer_thickness.jpg')
```

```python
# 壁面摩擦係数データの読み込み
data_cdf = pd.read_csv('output_cdf.csv', header=None)
```

```python
# 壁面摩擦係数の可視化
x_cdf = data_cdf[0]
y = data_cdf[1]

plt.plot(x_cdf, y)
plt.xlabel('distance (m)')
plt.ylabel('cdf')
plt.title('Visualization of CDF')
plt.savefig('cdf.jpg')
```
